```python
import csv
from pylab import *
f = open('cig.csv', 'rb')
reader = csv.reader(f)
xrow = []
yrow = []
for row in reader:
    string = row[0]
    string = string.split()
    header =[string[0]]
    a=map(float, string[1:])
    for x in a:
        header.append(x)
    xrow.append(header[1])
    yrow.append(header[2])
    print header
f.close()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('nicotin and tar')
ax.set_xlabel('tar')
ax.set_ylabel('nicotin')
ax.scatter(xrow, yrow)
fig.savefig("cig_scatter.png")
```
