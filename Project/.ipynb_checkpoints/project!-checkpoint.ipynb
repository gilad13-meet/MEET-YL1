```python
import csv
```

```python
f = open('physics.csv', 'rb')
reader = csv.reader(f)
count = 0
meters = []
time = []
speed = []
for row in reader:
    if count >0:
        string = row[0]
        string.replace("/t", " ")
        plist = string.split()
        plist = map(float, plist)
        time.append(plist[0])
        meters.append(plist[1])
        if count>1:
            speed.append(plist[2])
    count+=1
print speed
```

```python
from pylab import *
```

this graph describes who far is the cupcake paper from the sensor.

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('time -place graph')
ax.set_xlabel('time(s)')
ax.set_ylabel('place(m)')
ax.scatter(time, meters)
fig.savefig("time-place.png")
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('time -place graph')
ax.set_xlabel('time(s)')
ax.set_ylabel('place(m)')
ax.plot(time, meters)
fig.savefig("time-placeg.png")
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('time -speed graph')
ax.set_xlabel('time(s)')
ax.set_ylabel('speed(m/s)')
ax.scatter(time[1:], speed)
fig.savefig("time-speed.png")
```
